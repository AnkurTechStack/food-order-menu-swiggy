```python
#Creating a Food Order Menu for the restaurant partner of Swiggy

#Program Features-
#1.Add new items--->Category wise
#2.View all items
#3.Search Items
#4.Remove/Delete Items
#5.Show Statistics-No. of items, Avg. Price, Max. Price, Min. Price
#6.Exit
```

```python
menu = {'dal makhani': ('main course', 200.0),
        'panner': ('main course', 400.0),
        'chilly potato': ('starters', 300.0),
        'sushi': ('main course', 750.0),
        'malai kofta': ('main course', 450.0),
        'gulab jamun': ('desserts', 230.0)
       
       }
```

```python
#1. Adding new items--->Category wise
#Item name, category, price, to be displayed in the dictionary and a successful message.
def add_items():
    name=input("Enter Item Name- ").strip().lower()
    category=input("Enter the category (starters/main course/desserts)- ").strip().lower()
    price=float(input("Enter the price of the item-"))
    menu[name]=(category, price)
    print(f"\n\n{name.title()} added successfully")

#2. View all items
#Category Level-1 item with price followed by another item with price.
#Dictionary for category.
#Category wise list.
#Add item and price in the category.
#Print category wise values.
def view_menu():
    categorized={}
    for item, (category, price) in menu.items: 
        if category not in categorized:
            categorized[category]=[]
        categorized[category].append((item,price))
    for cat in categorized:
        print(f"\n\nCategory {cat.title()}")
        for item, price in categorized[cat]:
            print(f"- {item.title()} @ Rs.{price}")
    print()
    print()

#3. Searching Food items
#Item name, find inside the dictionary, and print item name in category of price x.
def search_items():
    keyword=input("Enter the Item Name to be search- ").strip().lower()
    found=False
    for item in menu:
        if keyword == item:
            cat,price = menu[item]
            print()
            print(f"\n\nFound {item.title()} of category-{cat} is priced at {price}\n\n")
            found=True
    if not found:
        print()
        print("\n\nItem not found\n\n")
                
        
   
#4. Removing/Deleting items
#Input of item name, which needs to be deleted.
#Search inside the dictionary.
#True-Delete.
#False-No item found.

def delete_items():
    name=input("Enter Item Name to be deleted- ").strip().lower()
    if name in menu:
        del menu[name]
        print()
        print()
        print(f"\n\n{name.title()} deleted successfully\n\n")
    else:
        print()
        print()
        print("\n\nNo such item found in the menu\n\n")

#5. Show Statistics-No. of items, Avg. Price, Max. Price, Min. Price.
def stats():
    price=[price for i, price in menu.values()]
    print("Statistics results as follows-\n")
    print("Total Items-", len(menu))
    print("Average Price-", round(sum(price)/ len(menu), 2))
    print("Cheapest Item-", min(price))
    print("Expensive Item-", max(price))

#Main Function
def main():
    while True:
        print("Title-Swiggy Food Menu Manager")
        print("1. Add New Item ")
        print("2. View all items ")
        print("3. Search Item ")
        print("4. Delete Item ")
        print("5. Statistics ")
        print("6. Exit ")

        choice=input("Please choose between 1-6: ")

        if choice=='1':
            add_items()
        elif choice=='2':
            view_menu()
        elif choice=='3':
            search_items()
        elif choice=='4':
            delete_items()
        elif choice=='5':
            stats()
        elif choice=='6':
            print("\n\nThank you for using the tool\n\n")
            break
        else:
            print("\n\n\nInvalid Output\n\n")
           
#Calling Main function
main()

```
